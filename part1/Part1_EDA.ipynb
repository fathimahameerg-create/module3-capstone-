```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("data/insurance.csv")
```

```python
df = df.drop_duplicates()
```

```python
for col in df.select_dtypes(include='object'):
    print(col)
    print(df[col].value_counts())
    print()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("plots/heatmap.png")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df["charges"])
plt.title("Boxplot of Charges")
plt.savefig("plots/charges_boxplot.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="sex", data=df)
plt.title("Gender Distribution")
plt.savefig("plots/gender_count.png")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="bmi", y="charges", data=df)
plt.title("BMI vs Charges")
plt.savefig("plots/bmi_vs_charges.png")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="age", y="charges", data=df)
plt.title("Age vs Charges")
plt.savefig("plots/age_vs_charges.png")
plt.show()
```

```python
sns.pairplot(df)
plt.savefig("plots/pairplot.png")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x="smoker", y="charges", data=df)
plt.title("Charges by Smoking Status")
plt.savefig("plots/smoker_vs_charges.png")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x="sex", y="charges", data=df)
plt.title("Charges by Gender")
plt.savefig("plots/gender_vs_charges.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="region", y="charges", data=df)
plt.title("Charges by Region")
plt.savefig("plots/region_vs_charges.png")
plt.show()
```
